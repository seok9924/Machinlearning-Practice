```python
a=X_train.columns[-1:]
y_train =X_train[a]
X_train.drop(['count'],axis=1,inplace=True)
#X_test.drop(['count'],axis=1,inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor


for i in (10,20,30,40,50,60,70,80,1000):
    rf_rg=RandomForestRegressor(n_estimators=i,random_state = 15) 
    rf_rg.fit(X_train,y_train)
    score= rf_rg.score(X_train,y_train)
    print(score)
    
```
